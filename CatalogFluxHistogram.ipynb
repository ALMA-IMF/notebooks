```python
from astropy import table
```

```python
def try_float(x):
    try:
        return float(x)
    except:
        return np.nan
def read_table(fn):
    with open(fn) as fh:
        lines = fh.readlines()
    newlines = []
    start = False
    for line in lines:
        if not start:
            if 'begin{tabular}' in line:
                newlines.append(line)
            if 'Core name' in line:
                newlines.append(line)
                start = True
                continue
        if start:
            if r'\\' in line and 'J2000' not in line and 'multicolumn' not in line:
                newlines.append(line.replace(r'\,','').replace("$^*$",""))
    #return newlines
    tb = ascii.read("\n".join(newlines), format='latex')
    tb['T'] = np.array([try_float(x) for x in tb['T']], dtype='float')
    tb['Mass'] = np.array([x.split(r'$\pm$')[0] if r'$\pm$' in x else try_float(x) for x in tb['Mass$^{\\,(1)}$']], dtype=float)
    tb['eMass'] = np.array([x.split(r'$\pm$')[1] if r'$\pm$' in x else try_float(x) for x in tb['Mass$^{\\,(1)}$']], dtype=float)
    tb['S_peak'] = np.array([x.split(r'$\pm$')[0] if r'$\pm$' in x else try_float(x) for x in tb['$S^{\\rm Peak}$']], dtype=float)
    tb['eS_peak'] = np.array([x.split(r'$\pm$')[1] if r'$\pm$' in x else try_float(x) for x in tb['$S^{\\rm Peak}$']], dtype=float)
    tb['S_int'] = np.array([x.split(r'$\pm$')[0] if r'$\pm$' in x else try_float(x) for x in tb['$S^{\\rm int}$']], dtype=float)
    tb['eS_int'] = np.array([x.split(r'$\pm$')[1] if r'$\pm$' in x else try_float(x) for x in tb['$S^{\\rm int}$']], dtype=float)
    for key in (r'$\alpha_{\rm BE}^{\,(1)}$', '$Size^{\\,(1)}$'):
        tb[key] = np.array([try_float(x) for x in tb[key]], dtype='float')
    tb['n'] = list(map(str, tb['n']))
    tb['PA'] = list(map(str, tb['PA']))
    tb['PA_1'] = list(map(str, tb['PA_1']))
    return tb
```

```python
lcats = {key: read_table(f'latex-smoothed/{key.split(".")[0]}-catpap.tex')
         for key in distances}
for key in distances:
    lcats[key]['distance'] = distances[key]*u.kpc

lconcat = table.vstack(list(lcats.values()))
```

```python
def theo_temperature(flux_3000au):
    m = (60-20) / (np.log10(4e3) - np.log10(10))
    b = 60 - m * np.log10(4e3)
    rslt = np.log10(flux_3000au.to(u.mJy).value) * m + b
    rslt[rslt < 20] = 20
    return rslt*u.K
```

```python
pl.plot(lconcat['S_peak'], lconcat['T'], '.')
pl.plot(lconcat['S_peak'], theo_temperature(lconcat['S_peak']*u.mJy), '.')
```

```python
from astropy.modeling.models import BlackBody
```

```python
def mass(flux, temperature, distance, kappa=0.1*(225/1000)**1.5*u.cm**2/u.g):
    #(distance**2 / kappa).to(u.M_sun)
    return (flux * distance**2 / (kappa * BlackBody(temperature)(225*u.GHz)*u.sr)).to(u.M_sun)
```

```python
pl.loglog(lconcat['Mass'], mass(lconcat['S_peak']*u.mJy,
                              theo_temperature(lconcat['S_peak']*u.mJy),
                              lconcat['distance'].quantity), '.')
pl.plot([0.1,150],[0.1,150])
pl.xlabel("Mass [Louvet]")
pl.ylabel("Mass [T(Robitaille 2017)]")
```
