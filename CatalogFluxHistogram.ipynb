```python
%matplotlib inline
from astropy import units as u
import pylab as pl
import numpy as np
pl.rcParams['figure.facecolor']='w'
```

```python
flxs = []
for key in cats:
    cat = cats[key]
    catmask = (cat['FG'] == 0) & (cat['FM03'] == 0)
    flx_b3 = (cat['FXT_BST03'][catmask]*u.Jy * (distances[key]*u.kpc)**2).to(u.Jy*u.kpc**2).value
    flxs.append(flx_b3)
pl.hist(flxs, histtype='stepfilled', stacked=True, bins=np.geomspace(0.002,10,25))
pl.semilogx()
pl.xlabel("3mm Luminosity (F$_\\nu$ d$^2$ [Jy kpc$^2$])")
```

```python
ttl = 0
ttlgood = 0
ttlsig = 0
ttlall = 0
for key in cats:
    cat = cats[key]
    catmask = (abs(cat['GOODM03']) > 1 ) & (abs(cat['SIGN'] > 1))
    ttlgood += (abs(cat['GOODM03']) > 1 ).sum()
    ttlsig += (abs(cat['SIGN'] > 1)).sum()
    ttl += catmask.sum()
    ttlall += len(cat)
    print(key, catmask.sum(), ttl, ttlgood, ttlsig, ttlall)
```

```python
pl.figure(figsize=(12,8))
pl.rcParams['font.size'] = 18
szs = []
for key in cats:
    cat = cats[key]
    catmask = (abs(cat['GOODM03']) > 1 ) & (abs(cat['SIGNM03'] > 1))
    sz_b3 = (cat['AFWHM01'][catmask]*u.arcsec * (distances[key]*u.kpc)).to(u.au, u.dimensionless_angles()).value
    szs.append(sz_b3)
pl.hist(szs, histtype='stepfilled', stacked=True, bins=np.geomspace(2.5e3,3e4), color=['darkgreen']*15,
       label="ALMA-IMF")
pl.legend(loc='best')
pl.semilogx()
pl.ylabel("Number of Sources")
pl.xlabel("Size [AU]")
pl.savefig("ALMA-IMF_sizes.png", bbox_inches='tight')
```
